```python
df = pd.read_csv("/content/amazon-consumer-behaviour-dataset/Amazon Customer Behavior Survey.csv")
```

```python
df.dropna(inplace=True)
df.pop("Timestamp") #drop the Timestamp column. not needed for analysis
```

```python
import matplotlib.pyplot as plt


def generate_frequency_report(df, col_name, title):
  freq_dict = {}

  unique_occurences_freq = df[col_name].value_counts() #get the unique values of that column and their frequency

  labels = [key for key,_ in unique_occurences_freq.items()]
  sizes = [val for _,val in unique_occurences_freq.items()]

  plt.pie(sizes, labels=labels, autopct="%1.1f%%")
  plt.title(title)
  plt.show()

  for occurence, freq in unique_occurences_freq.items():
    freq_dict[occurence] = freq


```

```python
cols_to_report = {
    "Gender":"Consumers Gender Distribution",
    "Purchase_Frequency":"Consumers Purchase Frequency Distribution",
    "Browsing_Frequency":"Consumers Browsing Frequency Distribution",
    "Product_Search_Method": "How Do Consumers Typically Search For Products",
    "Search_Result_Exploration": "How Do Consumers Tend To Explore Search Result Pages",
    "Customer_Reviews_Importance": "Consumer Ratings Distribution Of Customer Reviews",
    "Add_to_Cart_Browsing": "Do Consumers Add Products While Browsing On Amazon",
    "Cart_Completion_Frequency": "How Often Do Consumers Complete Purchase After Adding Products To Cart",
    "Cart_Abandonment_Factors": "Factors That Influence Consumers To Abandon A Purchase In Their Cart",
    "Saveforlater_Frequency": "How Often Do Consumers Use The Save For Later Feature",
    "Review_Left": "Do Consumers Leave Product Reviews",
    "Review_Reliability": "How Much Do Consumers Rely On Product Reviws When Making A Purchase",
    "Review_Helpfulness": "Do Cosnumers Find Helpful Information From Other Customer's Reviews",
    "Personalized_Recommendation_Frequency": "How Often Do Consumers Receive Personalized Recommendation",
    "Recommendation_Helpfulness": "Do Consumers Make Purchases Based On Recommendation",
    "Rating_Accuracy ": "Consumers Rating Of Accuracy And Relevancy Of The Recommendations They Received",
    "Shopping_Satisfaction": "Consumer Shopping Satisfaction Distribution",
    "Service_Appreciation": "Aspects Of Amazon's Services Consumers Appreciate The Most",
    "Improvement_Areas": "Areas Of Improvement"
}
```

```python
for col in cols_to_report:
  generate_frequency_report(df, col, cols_to_report[col])
```
